```python
import pandas as pd
import seaborn as sns; sns.set()
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('population.csv')
df.head()
```

```python
'''By Year'''
df['Year'].unique()
```

```python
sns.countplot(df['Year'])
plt.ylim(10000, 15000)
```

```python
sns.countplot(df['District.Code'])
```

```python
fig, ax = plt.subplots(figsize = (20, 20))
sns.barplot(x = 'Number',
           y = 'Age',
           data = df)
```

```python
df_birth = pd.read_csv('births.csv')
df_birth.head()
```

```python
genders = ['Boys', 'Girls']
a = len(df_birth[df_birth['Gender'] == 'Boys'])
b = len(df_birth[df_birth['Gender'] == 'Girls'])
y = np.array([a, b])
sns.barplot(genders, y)
plt.show()
print('Boys:', a)
print('Girls:', b)
```
