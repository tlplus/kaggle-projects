```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('players_20.csv')
df.head()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
df.columns
```

```python
df.drop(['sofifa_id', 'player_url', 'long_name'], axis = 1, inplace = True)
df.head()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (12, 6) )
sns.countplot(x = 'age',
             data = df)
print("Average age: ", round(df['age'].mean(), 1))
fig, ax = plt.subplots(1, 1, figsize = (12, 6))
sns.kdeplot(df['age'], shade = True)
plt.xlabel('age')
```

```python
youngestClubs = df.groupby(['club'])['age'].mean().sort_values(ascending = True).head(10)
oldestClubs = df.groupby(['club'])['age'].mean().sort_values(ascending = False).head(10)
print("Youngest", youngestClubs)
print("\nOldest", oldestClubs)
```

```python
wage = df.groupby(['club'])['wage_eur'].mean().sort_values(ascending = False).head(10)
wage = wage.reset_index()
wage.rename(columns = {'club': 'Club', 'wage_eur': 'Wage (Euros)'}, inplace = True)
fig, ax = plt.subplots(figsize = (12, 6))
sns.barplot(x = 'Club',
           y = 'Wage (Euros)',
            palette = 'husl',
           data = wage)
plt.xticks(rotation = 60)
```

```python
potential = df.groupby(['club'])['potential'].mean().sort_values(ascending = False).head(10)
potential
```
