```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('transfers.csv')
df.head()
```

```python
'''Which position is the most in this list?'''
fig, ax = plt.subplots(figsize = (30, 10))
sns.countplot(df['Position'])
```

```python
'''Centre Forward players are the most ones to be transfered '''
print('Total Centre Forward: ', len(df[df['Position'] == 'Centre-Forward']))
```

```python
'''Most youngest players to be transfered'''
df.sort_values(by = 'Age')[['Name', 'Age', 'Team_from', 'Season']].head(10)
```

```python
df['Team_from'].unique()
```

```python
'''Age distribution of few clubs who have transfered their players'''

fig, ax = plt.subplots(figsize = (10, 6))
clubs = ['Real Madrid', 'FC Barcelona', 'Man City', 'Man Utd',
        'Arsenal', 'Bayern Munich ']
df_age = df.loc[df['Team_from'].isin(clubs) & df['Age']]

sns.violinplot(x = 'Team_from',
              y = 'Age',
              data = df_age,
              size = 40)

plt.xlabel('Club')
```

```python
'''Age distribution of some clubs that bought news players'''

fig, ax = plt.subplots(figsize = (10, 6))
clubs = ['Real Madrid', 'FC Barcelona', 'Man City', 'Man Utd',
        'Arsenal', 'Bayern Munich ']
df_age = df.loc[df['Team_to'].isin(clubs) & df['Age']]

sns.violinplot(x = 'Team_to',
              y = 'Age',
              data = df_age,
              size = 40)

plt.xlabel('Club')
```

Arsenal looks like buying all range of ages

Real Madrid heavily bought players from 20 to 25

```python
'''Most expensive transfer ever'''

top_transfers = df.sort_values(by = 'Transfer_fee', ascending = False)[['Name', 'Team_to', 'Season', 'Transfer_fee']].head(10)
print(top_transfers)

plt.figure(figsize = (15, 6))
sns.barplot(x = 'Name',
           y = 'Transfer_fee',
           data = top_transfers,
           hue = 'Season')

tick_val = [50000000, 100000000, 200000000]
tick_lab = ['50m', '100m', '200m']

# Adapt the ticks on the x-axis
plt.yticks(tick_val, tick_lab)
plt.ylabel('Transfer Fee (Millions)')
plt.xlabel('Players')
plt.title('Players Transfer Fee')
plt.show()
```

Neymar to PSG has been the most expensive transfer

Neymar appeared twice in this list as he also transfered to Barcelona in 2013/14

```python
df['League_to'].unique()
```

```python
plt.figure(figsize = (15, 6))

leagues = ['LaLiga', 'Serie A', 'Ligue 1', 'Premier League', '1.Bundesliga']
league = []
for i in leagues:
    total = len(df[df['League_to'] == i])
    league.append(total)
    
sns.barplot(x = leagues,
           y = league)

plt.ylabel('Total Bought Players')
plt.xlabel('League')
```

Compared to other leagues, Premier League teams bought a lot of players.

```python
plt.figure(figsize = (30, 10))

real_madrid = df[df['Team_to'] == 'Real Madrid']
print(real_madrid['Position'].unique())

positions = ['Right Winger', 'Central Midfield', 'Defensive Midfield',
       'Second Striker', 'Goalkeeper', 'Left Midfield',
       'Attacking Midfield', 'Centre-Forward', 'Right Midfield',
       'Centre-Back', 'Left Winger', 'Right-Back', 'Left-Back']
total = []
for position in positions:
    total.append(len(real_madrid[real_madrid['Position'] == position]))
    
sns.barplot(x = positions,
           y = total,
           palette = 'summer')

plt.title("Real Madrid Totally Bought Players and Their Positions ", size = 30)
```

```python
sns.jointplot(x = 'Market_value',
             y = 'Transfer_fee',
             data = real_madrid)
tick_val = [50000000, 100000000, 200000000]
tick_lab = ['50m', '100m', '200m']

# Adapt the ticks on the x-axis
plt.yticks(tick_val, tick_lab)
plt.xlabel('Market Value')
plt.ylabel('Transfer Fee')
'''There are some missing value in market value'''
```
