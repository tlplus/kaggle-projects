```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
df = pd.read_csv('transfers.csv')
df.head()
```

```python
fig, ax = plt.subplots(figsize = (30, 10))
sns.countplot(df['Position'])
```

```python
'''Centre Forward players are the most ones to be transfered '''
print('Total Centre Forward: ', len(df[df['Position'] == 'Centre-Forward']))
```

```python
df.sort_values(by = 'Age')[['Name', 'Age', 'Team_from', 'Season']].head(10)
```

```python
df['Team_from'].unique()
```

```python
'''Age distribution of few clubs who have transfering players'''

fig, ax = plt.subplots(figsize = (10, 6))
clubs = ['Real Madrid', 'FC Barcelona', 'Man City', 'Man Utd',
        'Arsenal', 'Bayern Munich ']
df_age = df.loc[df['Team_from'].isin(clubs) & df['Age']]

sns.violinplot(x = 'Team_from',
              y = 'Age',
              data = df_age,
              size = 40)

plt.xlabel('Club')
```

```python
'''Age distribution of some clubs buying news players'''

fig, ax = plt.subplots(figsize = (10, 6))
clubs = ['Real Madrid', 'FC Barcelona', 'Man City', 'Man Utd',
        'Arsenal', 'Bayern Munich ']
df_age = df.loc[df['Team_to'].isin(clubs) & df['Age']]

sns.violinplot(x = 'Team_to',
              y = 'Age',
              data = df_age,
              size = 40)

plt.xlabel('Club')
```

Arsenal looks like buying all range of ages

Real Madrid heavily bought players from 20 to 25

```python
df.sort_values(by = 'Transfer_fee', ascending = False)[['Name', 'Team_to', 'Season', 'Transfer_fee']].head(10)
```

```python
df['League_to'].unique()
```

```python
plt.figure(figsize = (15, 6))

leagues = ['LaLiga', 'Serie A', 'Ligue 1', 'Premier League', '1.Bundesliga']
league = []
for i in leagues:
    total = len(df[df['League_to'] == i])
    league.append(total)
    
sns.barplot(x = leagues,
           y = league)
```

Premier League teams bought a lot of players compared to others
